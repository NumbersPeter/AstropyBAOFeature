
## SQL

```python
SELECT TOP 50000 -- 테스트용으로 5만 개만 먼저 추출
    objid, ra, dec, z, zerr, type
FROM Galaxy
WHERE 
    z BETWEEN 0.4 AND 0.7 -- 논문의 실패 요인을 보완하는 핵심 필터 
    AND zWarning = 0
    AND clean = 1
    AND (class = 'GALAXY')
```

```python
from astropy.cosmology import FlatLambdaCDM

# Planck 2018 결과를 참고한 표준 우주론 파라미터 설정
# 논문에서 언급한 우주 팽창 속도 계산의 기초가 됨 [cite: 55]
cosmo = FlatLambdaCDM(H0=67.7, Om0=0.31)

def get_comoving_distance(z):
    """적색 편이 값을 공변 거리(Mpc)로 변환"""
    return cosmo.comoving_distance(z).value
```
